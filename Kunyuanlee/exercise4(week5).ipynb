2. Write a python function that output the following formatted string based on the input iterable:
'The n numbers are: i1, i2, i3, ...'
where n is the number in the iterable and i1, i2 ... are the items inside the iterable.
for example:
func([1, 2, 3]) will return 'The 3 numbers are: 1, 2, 3'
func((6, 7, 8, 9)) will return 'The 4 numbers are: 6, 7, 8, 9'

```python
def func(seq):
    return "The "+str(len(seq))+" numbers are: "+str(seq)[1:-1]

# Testing
print(func([1, 2, 3]))
print(func((6, 7, 8, 9)))
```

6. Write a function that will format each item in an input iterable object into a customized width
syntax: func(iterable, width)
for example:
func([1, 2, 3, 4, 5], 3) return '  1  2  3  4  5'
func([1, 2, 3, 4, 5], 2) return ' 1 2 3 4 5'

```python
def func(iterable, width):
  
    res = ""
  
    for each_item in iterable:
        res = res + "{:>{width}}".format(each_item, width = width)
  
    return res


print(func([1, 2, 3, 4, 5], 3))
print(func([1, 2, 3, 4, 5], 2)) 
```

```python
# 1) generate the url for a given list of gage numbers; 
from urllib.request import urlopen
import math
import datetime

def generate_urlink(gagelist, begin_date, end_date):
    try:
        bdate = datetime.date.fromisoformat(begin_date)
        edate = datetime.date.fromisoformat(end_date)
    except:
        raise ValueError ("The input formats for the begin_date and end_date must be YYYY-MM-DD")
        
    gages = ('&site_no={}' * len(gagelist)).format(*gagelist)
    period = f'&period=&begin_date={begin_date}&end_date={end_date}'
    return f'https://waterdata.usgs.gov/nwis/dv?&cb_00060=on&format=rdb{gages}&referred_module=sw{period}'

print('The follow url is created: ' + generate_urlink(['06803495', '06803486'], '2000-10-10', '2019-09-30'))
```

```python
#2) read this url and download the data to csv file (do not use numpy or pandas); 
def read_usgs_flowdata(url):
    f = urlopen(url)
    return f.readlines()

lines = read_usgs_flowdata(generate_urlink(['06803495', '06803486'], '2000-10-10', '2019-09-30'))

#Testing if the program reads
for r in lines:
    print(r)
```

```python
#This section cleans the messey data and replace \t with , 
def save_date(data_lines, csvfile):
    for n, l in enumerate(data_lines):
        if not l.startswith(b'#'):
            break
    
    # now start working on the lines containing data
    with open(csvfile, 'w') as f:
        # extract and write headers
        f.write(data_lines[n].decode().replace('\t', ','))
        
        # skip the column spec line
        n += 2
        for i, l in enumerate(data_lines[n:]):
            if l.startswith(b'USGS'):
                f.write(l.decode().replace('\t', ','))
                
save_date(lines, 'usgs_streamflow.csv') 

#testing the 10 data gaguge of begin date and end date
! head -10 'usgs_streamflow.csv'
print('.')
print('.')
print('.')
! tail -10 'usgs_streamflow.csv'
```

```python
#3) based on the downloaded data calculate the monthly stasticstics including maximum,
#minimum and average of the streamflow data and save it to another csv file. 
#The data need to be formatted to 2 decimal digits. 
import math
def monthlyStats(myStreamflowData):
    site = ''
    date = ''
    begin = True
    with open(myStreamflowData) as f:
        # skip header
        f.readline()
        # f.readline()
        
        siteIdx = 1
        dateIdx = 2
        valueIdx = 3
        
        for j, dataLine in enumerate(f):
            dataItems = dataLine.split(',')
            # update key when site or month changes
            # note: 7th item in date is the month
            # if current 'date' does not equal previous date or
            # if current 'site' does not equal previous site
            # then reset date and site
            if dataItems[dateIdx][:7] != date or dataItems[siteIdx] != site:
                site = dataItems[siteIdx]
                date = dataItems[dateIdx][:7]
                key = f'{site},{date}'
                if begin:
                    siteMonth = {key:[]}
                    begin = False
                else:
                    siteMonth[key] = []
                
            if dataItems[valueIdx].replace('.','0').isnumeric():
                siteMonth[key].append(float(dataItems[valueIdx]))
            else:
                # if value is 'ice' then print the dataItems
                print(dataItems)


                
    # write data to a new file
    with open('monthly_csv', 'w') as fw:
        # add header
        fw.write('siteNumber,month,min,max,mean\n')
        for i in siteMonth:
            if siteMonth[i] == []:
                print(i, siteMonth[i])
            else:
                minVal = min(siteMonth[i])
                maxVal = max(siteMonth[i])
              
                meanVal = sum(siteMonth[i])/len(siteMonth[i])
                
               
                fw.write(f'{i}-01,{minVal:.2f},{maxVal:.2f},{meanVal:.2f}\n')


monthlyStats('monthly_streamflow.csv')

import csv
 
with open('monthly_streamflow.csv', newline='') as File:  
    reader = csv.reader(File)
    for row in reader:
        print(row)

```

```python
  #4) calculate the average annual runoff of each gage (expressed in acre-feet) 
def annualRunoff(myStreamflowData):
    currentSite = ''
    currentYear = ''
    begin = True
    with open(myStreamflowData) as f:
        f.readline()
        siteIdx = 1
        dateIdx = 2
        valueIdx = 3
        for k, dataLine in enumerate(f):
            dataItems = dataLine.split(',')
            
            if dataItems[dateIdx][:4] != currentYear or dataItems[siteIdx] != currentSite:
                currentYear = dataItems[dateIdx][:4]
                currentSite = dataItems[siteIdx]
                key = f'{currentSite},{currentYear}'
                if begin:
                    siteYear = {key:[]}
                    begin = False
                else:
                    siteYear[key] = []
            
            if dataItems[valueIdx].replace('.','0').isnumeric():
                
                #convert from 'cubic feet per second' to 'acre feet per day'
                acreFeet = float(dataItems[valueIdx])*24*60*60/43560
                siteYear[key].append(float(acreFeet))
                

    with open('annualRunoff.csv', 'w') as fw:
        fw.write('siteNumber,  year,  annual runoff (ac/ft)\n')
        for key in siteYear:
            totalRunoff = sum(siteYear[key])
            fw.write(f'{key}-01-01,{totalRunoff:.2f}\n')
        

annualRunoff('usgs_streamflow.csv')


import csv
 
with open('annual_runoff.csv', newline='') as File:  
    reader = csv.reader(File)
    for row in reader:
        print(row)

```
